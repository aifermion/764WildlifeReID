```python
class FeralCatsAkl(datasets.DatasetFactory):
    def create_catalogue(self) -> pd.DataFrame:
        df = pd.DataFrame({
            'image_id': feral_cats['image_id'],
            'identity': feral_cats['identity'],
            'path': feral_cats['path'],
        })

        # df['path'].apply(lambda x: dataset_path + x)

        return df
```
