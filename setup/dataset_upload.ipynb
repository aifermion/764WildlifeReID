In this notebook we test whether our new datasets were created correctly

```python
import pandas as pd
from wildlife_datasets import datasets
```

```python
# for checking whether the paths are readable
from PIL import Image
import matplotlib.pyplot as plt

def check_image(path):
    try:
        image_path = path
        image = Image.open(image_path)  

        # Display the image
        plt.imshow(image)
        plt.axis('off')  # Hide the axis
        plt.show()
    except:
        return False
```

## Baseline Check

```python
dataset_path = "../data/processed/cropped"
fca_base = datasets.FeralCatsAkl(dataset_path).create_catalogue()
check_image(fca_base.iloc[0].path)
```

## Maxim-augmented Check

```python
maxim_dataset_path = '/Users/fmb/GitHub/764WildlifeReID/data/formatted_datasets/FeralCatsAkl_maxim'
fca_maxim = datasets.datasets.FeralCatsAkl_Maxim(maxim_dataset_path).create_catalogue()
check_image(fca_maxim.iloc[0].path)
```

## HIDiff-augmented Check

```python
hidiff_dataset_path = '/Users/fmb/GitHub/764WildlifeReID/data/formatted_datasets/FeralCatsAkl_hidiff'
fca_hidiff = datasets.datasets.FeralCatsAkl_HIDiff(hidiff_dataset_path).create_catalogue()
check_image(fca_hidiff.iloc[0].path)
```

## SRMNet-augmented Check

```python
srmnet_dataset_path = '/Users/fmb/GitHub/764WildlifeReID/data/formatted_datasets/FeralCatsAkl_srmnet'
fca_srmnet = datasets.datasets.FeralCatsAkl_SRMNet(srmnet_dataset_path).create_catalogue()
check_image(fca_srmnet.iloc[0].path)
```
